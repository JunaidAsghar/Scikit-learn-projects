# Neural Network Classifier

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

cancer = load_breast_cancer()
X_train,X_test, y_train,y_test = train_test_split(cancer.data, cancer.target, random_state= 0)

mlp = MLPClassifier(random_state=42)
mlp.fit(X_train,y_train)

print("Accucracy on the training set : {0:.2}".format(mlp.score(X_train,y_train)))
print("Accucracy on the test set : {0:.2}".format(mlp.score(X_test,y_test)))
```

```python
# nomalizing the data to increase the accuracy and give the maximum iteration number to MLP

print ("Maximun Value at each column:\n{}". format (cancer.data.max(axis=0)))
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit(X_train).transform(X_train)
X_test_scaled = scaler.fit(X_test).transform(X_test)

MLP = MLPClassifier(max_iter=1000, random_state=42)
MLP.fit(X_train_scaled, y_train)


print("Accucracy on the training set : {0:.2}".format(MLP.score(X_train_scaled,y_train)))
print("Accucracy on the test set : {0:.2}".format(MLP.score(X_test_scaled,y_test)))

```
